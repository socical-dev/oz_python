#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

10번만에 맞히지 못하면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]

# 행맨 그림
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
]

class HangmanGame:
    def __init__(self):
        """
        게임 초기화 메서드.
        - 게임 시작 시 환영 메시지 출력
        - 최대 시도 횟수 설정 (10번)
        - 초기 시도 횟수 설정
        - 비어있는 단어와 추측한 문자들을 저장할 변수 설정
        """
        print("Hangman Game에 오신 것을 환영합니다.\n단어를 맞추시면 됩니다.\n(10번 안에 못 맞추시면 죽습니다.)")
        self.max_attempts = 10  # 기회는 총 10번
        self.attempts = 0       # 현재 시도 횟수
        self.secret_word = ""   # 정답 단어
        self.display_word = ""  # 유저에게 보여질 단어
        self.guessed_letters = []  # 이미 추측한 글자들

    def display(self):
        """
        현재 게임 상태를 출력하는 메서드.
        - 현재까지 맞힌 단어를 출력
        - 시도 횟수에 맞는 행맨 그림을 출력
        """
        print(" ".join(self.display_word))  # 현재 맞힌 글자 출력
        print(hangman_pics[self.attempts])  # 시도 횟수에 맞는 행맨 그림 출력

    def play(self):
        """
        게임을 실행하는 메서드.
        - 랜덤으로 단어 선택
        - 유저가 글자를 입력하고 맞추는 로직 실행
        - 시도 횟수에 맞춰 게임 진행
        """
        # 단어 랜덤 선택
        self.secret_word = random.choice(words)  # 단어 목록에서 랜덤으로 단어 선택
        self.display_word = ["_"] * len(self.secret_word)  # 단어 길이에 맞춰 밑줄로 초기화
        
        # 게임 시작
        while self.attempts < self.max_attempts:  # 최대 시도 횟수까지만 게임 진행
            self.display()  # 현재 상태와 그림을 출력

            # 사용자로부터 추측할 알파벳 입력 받기
            input_alph = input("있을 거 같은 영어 알파벳 하나를 입력해주세요!! : ").lower()

            # 이미 추측한 글자는 다시 입력할 수 없음
            if input_alph in self.guessed_letters:
                print(f"이미 입력한 알파벳입니다: {input_alph}")
                continue  # 다시 입력 받기

            # 입력값이 단어에 포함되어 있는지 확인
            if input_alph in self.secret_word:
                # 알파벳이 있으면 단어에 해당 위치 채워넣기
                for idx, letter in enumerate(self.secret_word):  # 단어의 각 글자와 비교
                    if letter == input_alph:
                        self.display_word[idx] = input_alph  # 맞힌 위치에 알파벳 채우기
                print("단어를 맞추셨군요. 더 힘내봐요")
            else:
                # 틀린 경우 시도 횟수 증가
                self.attempts += 1
                print(f"틀렸습니다. 남은 시도 횟수: {self.max_attempts - self.attempts}")

            # 추측한 글자를 이미 추측한 목록에 추가
            self.guessed_letters.append(input_alph)

            # 단어를 다 맞혔으면 게임 종료
            if "_" not in self.display_word:  # 모든 글자를 맞췄으면
                print(f"축하합니다! 단어를 맞추셨습니다. 시도 횟수: {self.attempts}")
                break
        else:
            # 시도 횟수 초과 시
            print(f"실패! 단어는 '{self.secret_word}'였습니다.")
            self.display()  # 마지막 행맨 그림 출력

if __name__ == "__main__":
    """
    게임을 시작하는 코드.
    - 게임 인스턴스를 생성하고 게임을 시작함
    """
    game = HangmanGame()  # 게임 객체 생성
    game.play()  # 게임 실행
```
